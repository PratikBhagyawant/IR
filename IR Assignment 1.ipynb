**Problem Statement**: Write a program for pre-processing of a text document such as stop word removal, stemming.

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
document="Information Retrieval and Natural Language Processing"
document
```

```python
Lower = document.lower()
Lower
```

```python
token = word_tokenize(Lower)
print(token)
```

```python
remove_punc=[i for i in token if(i.isalnum())]
print(remove_punc)
```

```python
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```python
remove_stopwords=[i for i in remove_punc if i not in stop_words]
print(remove_stopwords)
```

```python
pos_tagged_tokens = nltk.pos_tag(remove_stopwords)
print(pos_tagged_tokens)
```

```python
ps=PorterStemmer()
stemm_word=[]
for word in remove_stopwords:
    stemm_word.append(ps.stem(word))

print("text after stemming",stemm_word)  
```

```python
lemmatizer=WordNetLemmatizer()
lemm_word=[]
for word in remove_stopwords:
    lemm_word.append(lemmatizer.lemmatize(word))
print("text after lemmatization",lemm_word)  
```
