**Problem Statement**: Build the web crawler to pull product information and links from an e-commerce website.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def crawl(url):
    headers={
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36"
    }

    response =requests.get(url,headers=headers)
    soup=BeautifulSoup(response.text,'html.parser')

    product=[]
    price=[]
    ratings=[]

    p_name=soup.find_all('div',class_="KzDlHZ")
    p_price=soup.find_all('div',class_="Nx9bqj _4b5DiR")
    p_rating=soup.find_all('span',class_="Wphh3N")

    for i in p_name:
        product.append(i.text)

    for i in p_price:
        price.append(i.text)

    for i in p_rating:
        ratings.append(i.text)

    min_len=min(len(product),len(price),len(ratings))
    product=product[:min_len]
    price=price[:min_len]
    ratings=ratings[:min_len]

    df=pd.DataFrame({'Product':product,'Price':price,'Ratings':ratings,})
    return df
    
```

```python
df=crawl("https://www.flipkart.com/search?q=Laptop%20under%20200000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
```

```python
def image_crawler(url):
    head = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36"}
    response = requests.get(url,headers = head)
    soup = BeautifulSoup(response.text,'html.parser')

    image = []
    im_data = soup.find_all('img',class_="DByuf4")

    for i in im_data:
        image.append(i.get('src'))
    return image
```

```python
img=image_crawler("https://www.flipkart.com/search?q=Laptop%20under%20200000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")
```
